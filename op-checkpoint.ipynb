```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/luigg/Desktop/onepiece/ONEPIECE.csv')

df = df.rename(columns={'episode': 'Episode'})
df.head()
```

```python
df.drop(['Unnamed: 0','season','rank','trend'], axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df['total_votes'] = df['total_votes'].str.replace(',', '')
df['total_votes'] = df['total_votes'].astype(float)
```

```python
df.describe()
```

```python
df_rating_sorted = df.sort_values('average_rating',ascending=False)
df_rating_sorted.drop('total_votes', axis=1,inplace=True)
df_rating_sorted.head()
```

```python
df_rating_sorted.iloc[-1]
df_rating_sorted.tail()
```

```python
df_mean = df.groupby('start')['average_rating'].mean().reset_index()
```

```python
plt.plot(df_mean['start'], df_mean['average_rating'])
plt.title('Average rating vs year')
plt.xlabel('Year')
plt.ylabel('Average rating')
plt.show()
```

```python
df_total_votes_year = df.groupby('start')['total_votes'].sum().reset_index()
df_total_votes_year.head()
```

```python
plt.bar(df_total_votes_year['start'], df_total_votes_year['total_votes'])
plt.title('Number of votes per year')
plt.xlabel('Year')
plt.ylabel('Total votes')
plt.show()
```

```python
year_counts = df['start'].value_counts(sort=False).reset_index()
year_counts = year_counts.rename(columns={'start': 'num_ep_year'})
year_counts = year_counts.rename(columns={'index': 'start'})
```

```python
df_votes_year_ep = pd.merge(year_counts, df_total_votes_year, on= 'start')
df_votes_year_ep.head()
```

```python
plt.bar(df_votes_year_ep['start'],df_votes_year_ep['total_votes']/df_votes_year_ep['num_ep_year'])
plt.title('Mean of votes per year')
plt.xlabel('Year')
plt.ylabel('Total votes')
plt.show()
```

```python
df_votes_sorted =df.sort_values('total_votes',ascending=False)
df_votes_sorted.head()
```

```python


df2 = pd.read_csv('C:/Users/luigg/Desktop/onepiece/OnePieceratings.csv')
df2.drop('Season', axis=1,inplace=True)
df.drop('name', axis=1,inplace=True)
df2 = df2.rename(columns={'IMDB Rating': 'IMDB Rating(newer)'})
df2 = df2.rename(columns={'Year of Release': 'start'})
df_merged = pd.merge(df, df2, on=['Episode','start'])
df_merged = df_merged.rename(columns={'average_rating': 'IMDB Rating(older)'})
df_merged.drop('total_votes', axis=1,inplace=True)

```

```python
column_order = ['Episode', 'Title', 'start','IMDB Rating(older)','IMDB Rating(newer)']
df_merged = df_merged[column_order]
df_merged.tail()
```

```python
df_merged = pd.concat([df_merged, df2], ignore_index=True)
df_merged.tail()
```

```python
df_new_mean = df_merged.groupby('start')['IMDB Rating(newer)'].mean().reset_index()
```

```python
plt.plot(df_new_mean['start'], df_new_mean['IMDB Rating(newer)'],label='Newer')
plt.plot(df_mean['start'], df_mean['average_rating'],label='Older')
plt.title('Average rating vs year')
plt.xlabel('Year')
plt.ylabel('Average rating')
plt.legend()
plt.show()
```

```python
df_merged = df_merged.sort_values('IMDB Rating(newer)', ascending=False)
df_merged.drop('IMDB Rating(older)',axis=1,inplace=True)
df_merged.head(10)
```
